## 1. Применим метод k-means для данных diabetes.csv 
## Построим кластера для следующих комбинаций атрибутов   
## - показатель BMI, возраст Age 
## - показатель BloodPressure, возраст Age  
## Построим визуализацию результатов

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


df = pd.read_csv('diabetes.csv')

df = df.dropna()

df.head(10)
```

```python
# k-means для BMI, Age
X1 = df[['BMI', 'Age']].values
kmeans1 = KMeans(n_clusters=2, n_init=10, random_state=0).fit(X1)

plt.figure(figsize=(8, 6))
plt.scatter(X1[:, 0], X1[:, 1], c=kmeans1.labels_)
plt.scatter(kmeans1.cluster_centers_[:, 0], kmeans1.cluster_centers_[:, 1], s=300, c='red')
plt.xlabel('BMI')
plt.ylabel('Age')
plt.title('Кластеры для BMI and Age')
plt.show()

# k-means для BloodPressure, Age
X2 = df[['BloodPressure', 'Age']].values
kmeans2 = KMeans(n_clusters=2, n_init=10, random_state=0).fit(X2)

plt.figure(figsize=(8, 6))
plt.scatter(X2[:, 0], X2[:, 1], c=kmeans2.labels_)
plt.scatter(kmeans2.cluster_centers_[:, 0], kmeans2.cluster_centers_[:, 1], s=300, c='red')
plt.xlabel('BloodPressure')
plt.ylabel('Age')
plt.title('Кластеры для BloodPressure and Age')
plt.show()
```

## 2. Применим метод иерархической кластеризации с выводом дендрограммы для техже комбинаций атрибутов 

```python
#single

from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Для "BMI, Age"
linked1 = linkage(df[['BMI', 'Age']], 'single')

plt.figure(figsize=(10, 7))
dendrogram(linked1,
            orientation='top',
            labels=df.index.to_list(),
            distance_sort='descending',
            show_leaf_counts=True)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.title('Дендрограмма иерархической кластеризации для BMI and Age')
plt.show()

# Для "BloodPressure, Age"
linked2 = linkage(df[['BloodPressure', 'Age']], 'single')

plt.figure(figsize=(10, 7))
dendrogram(linked2,
            orientation='top',
            labels=df.index.to_list(),
            distance_sort='descending',
            show_leaf_counts=True)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.title('Дендрограмма иерархической кластеризации для BloodPressure и Age')
plt.show()
```

```python
#complete

# Для "BMI, Age"
linked1 = linkage(df[['BMI', 'Age']], 'complete')

plt.figure(figsize=(10, 7))
dendrogram(linked1,
            orientation='top',
            labels=df.index.to_list(),
            distance_sort='descending',
            show_leaf_counts=True)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.title('Дендрограмма иерархической кластеризации для BMI и Age')
plt.show()

# Для "BloodPressure, Age"
linked2 = linkage(df[['BloodPressure', 'Age']], 'complete')

plt.figure(figsize=(10, 7))
dendrogram(linked2,
            orientation='top',
            labels=df.index.to_list(),
            distance_sort='descending',
            show_leaf_counts=True)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.title('Дендрограмма иерархической кластеризации для BloodPressure и Age')
plt.show()
```
